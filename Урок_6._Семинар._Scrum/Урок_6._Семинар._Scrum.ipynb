# Урок 6. Семинар. Scrum
Задача:

В вашем распоряжении 3-х комнатная квартира, в которой на протяжении последних 20-ти лет не проводилось никаких ремонтных работ. Вы вместе со своими друзьями хотите самостоятельно сделать в ней современный ремонт, применив Scrum-фреймворк к работе над таким продуктом.

1.Опишите, как Вы организуете работу над данным продуктом в соответствии с фреймворком Scrum.

2.Сформируйте Product Backlog.

3.Разделите все работы на спринты, как Вы это видите.

4.Сформируйте Sprint Backlog любого одного спринта, который получился в результате деления работ на спринты.

5.Нарисуйте Scrum-доску и отразите на ней организацию работы над задачами этого спринта.

### Организация работы с помощью Scrum-фреймворка

1. Организация работы над продуктом в соответствии с фреймворком Scrum:

- Scrum Team: Сформируйте команду, включающую в себя Scrum Master (организатор процесса), Product Owner (отвечает за видение проекта и приоритеты) и команду разработчиков (выполняют работы).
  
- Product Backlog: Определите список всех требований к ремонту квартиры, упорядоченных по приоритетности.
  
- Sprint Planning: Проведите планирование спринтов, на котором команда выбирает элементы из Product Backlog и формирует Sprint Backlog.

- Daily Scrum: Ежедневные короткие собрания для отслеживания прогресса и координации действий команды.

- Sprint Review: По окончании спринта проводите обзор проделанной работы, демонстрируйте результаты и собирайте обратную связь.

- Sprint Retrospective: После каждого спринта команда обсуждает, что было сделано хорошо, что можно улучшить и какие действия должны быть предприняты в следующем спринте.

2. Формирование Product Backlog:

- Снос старых конструкций и мусора.
- Замена электропроводки.
- Замена труб и сантехники.
- Выравнивание стен и потолков.
- Укладка полов.
- Покраска стен и потолков.
- Установка дверей и окон.
- Монтаж мебели и оборудования.
- Декорирование и финальная отделка.

3. Разделение работ на спринты:

- Спринт 1: Подготовительные работы. Снос, вынос мусора, подготовка инструментов.
- Спринт 2: Замена электропроводки и труб, начало штукатурных работ.
- Спринт 3: Выравнивание стен и потолков, укладка полов.
- Спринт 4: Покраска стен, установка окон и дверей.
- Спринт 5: Монтаж мебели, оборудования и декорирование.

4. Сформировать Sprint Backlog для Спринта 1:

- Список комнат для сноса.
- План выноса мусора.
- Закупка необходимых инструментов.
- Подготовка квартиры к ремонтным работам.

5. Нарисовать Scrum-доску для Спринта 1:

| Задачи (To Do)                 | В процессе (In Progress) | Проверка (Review) | Готово (Done) |
|--------------------------------|--------------------------|-------------------|---------------|
| Закупить инструменты           |                          |                   |               |
| Список комнат для сноса        |                          |                   |               |
| Запланировать вывоз мусора     |                          |                   |               |
| Подготовить квартиру к работам |                          |                   |               |
