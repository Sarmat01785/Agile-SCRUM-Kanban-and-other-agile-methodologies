# Урок 2. Семинар. Гибкие методологии. История возникновения

Задача 

Вы с друзьями хотите разработать умный электросамокат. В рамках своего проекта вы хотите создать электросамокат, который будет работать на солнечной батарейке, иметь встроенный GPS-модуль, навигатор с картой города и отпечаток пальцев для активации самого электросамоката. На начальном этапе у вас не хватает достаточного количества денег, и вы привлекаете инвестора. Инвестор хочет понять, стоит ли инвестировать в вашу бизнес-идею и просит показать прототип такого электросамоката, а также хочет, чтобы в дальнейшем вы проводили разработку продукта по итеративной модели.

Опишите, каким образом вы организуете процесс разработки своего продукта, применив итеративную модель.

| № | Этап | Описание                                                                                       |
|---|------|-----------------------------------------------------------------------------------------------|
| 1 | Планирование | Начнем с определения целей и задач проекта, а также определения основных функций нашего электросамоката. Это поможет определить приоритеты и спланировать этапы работы. |
| 2 | Спринты      | Разобьем проект на короткие периоды времени, называемые спринтами. Каждый спринт будет длиться несколько недель, и в течение этого времени будем работать над определенными функциями продукта. |
| 3 | Сбор требований | Соберем все требования к продукту от инвестора и наших пользователей. Это поможет понять, какие функции наиболее важны и должны быть реализованы в первую очередь. |
| 4 | Разработка    | В течение каждого спринта будем разрабатывать определенные функции продукта. Будем работать в команде, используя методы парного программирования и ежедневные встречи для обсуждения прогресса и решения возникающих проблем. |
| 5 | Тестирование  | После завершения разработки каждой функции проведем тестирование, чтобы убедиться, что она работает корректно. Это поможет быстро обнаруживать и исправлять ошибки. |
| 6 | Демонстрация  | В конце каждого спринта продемонстрируем инвестору и пользователям результаты нашей работы. Это позволит им увидеть прогресс и дать обратную связь. |
| 7 | Рефакторинг   | После каждого спринта проанализируем код и внесем необходимые улучшения, чтобы сделать его более чистым и эффективным. |
| 8 | Ретроспектива  | В конце проекта проведем ретроспективу, чтобы обсудить, что прошло хорошо, а что можно улучшить в будущем. |
