# Гибкие методологии (Agile, SCRUM, Канбан и другие)


### Урок 12. Семинар. Как выбрать гибкую методологию и не ошибиться

#### Задача:

Вы находитесь в команде компании-разработчика, которую привлекли для реализации проекта по созданию мобильного приложения для крупной логистической компании. В рамках проекта Вам предстоит создать мобильное приложение, с помощью которого клиенты компании будут создавать заказы на перевозку грузов. У заказчика есть следующие требования:

1. Рассчитать бюджет всего проекта в самом начале – это необходимо для корректного бюджетирования деятельности компании. Рассчитать бюджет можно примерно, оценка бюджета может уточняться по мере разработки продукта.

2. Заказчик готов участвовать в процессе разработки.

3. У заказчика есть только идея, но нет чёткого представления, как сформировать техническое задание на разработку – он хочет обсудить это с Вами.

- Проанализируйте данный кейс и предложите наиболее подходящую модель разработки. Обоснуйте свой выбор.

### Решение: Выбор модели разработки

Для реализации проекта по созданию мобильного приложения для логистической компании предлагаю использовать гибкую методологию разработки, в частности, `Agile` с использованием `Scrum` или `Kanban`, в зависимости от предпочтений команды и динамики проекта. 

#### Обоснование выбора:

1. Предварительная оценка бюджета:

   - `Agile` позволяет провести итерационную оценку бюджета, начиная с высокоуровневых оценок и уточняя их с каждым спринтом.
   - Можно использовать метод `"Planning Poker"` для оценки трудозатрат по задачам и, исходя из этого, формировать бюджет проекта.

2. Участие заказчика в процессе:
   - `Agile` подразумевает тесное взаимодействие с заказчиком, что позволяет получать регулярную обратную связь и вносить корректировки в проект.
   - В `Scrum` заказчик может принимать участие в роли `Product Owner`, что позволяет ему напрямую влиять на приоритизацию задач.

3. Разработка технического задания:
   - В `Agile` техническое задание формируется постепенно, начиная с создания пользовательских историй и продуктового бэклога.
   - Работа ведётся в условиях неопределённости и изменяющихся требований, что идеально подходит для ситуации, когда у заказчика есть только начальная идея.

#### Предложенная структура проекта:

## Фазы проекта

### Фаза 1: Планирование и оценка
- Встреча с заказчиком для определения целей и приоритетов.
- Создание начального продуктового бэклога.
- Оценка бэклога с помощью методов `Agile` оценки (например, `Planning Poker`).
- Предварительное бюджетирование проекта на основе оценок.

### Фаза 2: Разработка и релизы
- Организация работы в спринтах (для `Scrum`) или с непрерывным потоком задач (для `Kanban`).
- Регулярные встречи (`Daily Stand-ups, Sprint Planning, Sprint Review, Sprint Retrospective`).
- Демонстрация прототипов заказчику и уточнение требований.
- Оперативное уточнение бюджета на основе реального прогресса.

### Фаза 3: Тестирование и деплой
- Проведение регулярного тестирования разработанных функций.
- Непрерывная интеграция и доставка (CI/CD) для быстрого получения обратной связи.
- Подготовка к запуску финальной версии приложения.

### Фаза 4: Поддержка и развитие
- Сбор и анализ пользовательских отзывов.
- Планирование итераций по улучшению и развитию продукта.
- Постоянная поддержка и оптимизация приложения.

## Роли в команде

- **Product Owner**: Представляет интересы заказчика, управляет продуктовым бэклогом.
- **Scrum Master (для Scrum)**: Фасилитирует процесс, помогает команде следовать выбранной методологии.
- **Команда разработки**: Работает над созданием продукта, состоит из программистов, дизайнеров, тестировщиков и т.д.

Используя `Agile`, команда сможет гибко подходить к изменениям требований и бюджета, обеспечивая эффективное взаимодействие с заказчиком и постоянное улучшение качества продукта.
