# Урок 8. Семинар. Lean

### Задача:

Проанализируйте свою личную повседневную работу в компании или работу одного из отделов в вашей компании на предмет наличия потерь.

1.Нарисуйте VSM-карту.

2.Найдите потери и классифицируйте их в соответствии с изученным на лекции материалом.

3.Подумайте, как можно исключить эти потери, и нарисуйте новую VSM-карту в соответствии с вашими предложениями.

### Решение:

Пример анализа и создания VSM:

Шаг 1: Создание текущей VSM-карты

Анализ отдела продаж. Текущая VSM-карта:

[Запрос клиента] → [Прием заказа] → [Обработка заказа] → [Подтверждение заказа] → [Доставка]


- Запрос клиента: Клиент отправляет запрос.
- Прием заказа: Сотрудник принимает и регистрирует заказ.
- Обработка заказа: Заказ обрабатывается в системе.
- Подтверждение заказа: Заказ проверяется и подтверждается клиенту.
- Доставка: Продукт доставляется клиенту.

Шаг 2: Определение потерь

Потери могут быть классифицированы следующим образом:

- Избыточные запасы: Заказы, ожидающие обработки.
- Переработка: Повторная регистрация заказа из-за ошибок.
- Ожидание: Время, затрачиваемое сотрудниками на ожидание следующего этапа.
- Неоптимальные процессы: Дублирование данных между системами.
- Дефекты: Ошибки в заказах, требующие дополнительной обработки.
- Перемещение: Неэффективная внутренняя логистика.
- Излишняя обработка: Ненужные шаги в процессе обработки заказов.

Шаг 3: Исключение потерь и создание новой VSM-карты

Предложения по исключению потерь:

- Автоматизация: Внедрите систему CRM для автоматического приема и обработки заказов.
- Обучение сотрудников: Повысьте квалификацию работников для снижения ошибок.
- Стандартизация процессов: Разработка четких инструкции по обработке заказов.
- Улучшение внутренней логистики: Оптимизация перемещение документов и товаров внутри компании.

Новая VSM-карта с учетом предложений:

[Запрос клиента] → [Автоматический прием заказа] → [Автоматическая обработка заказа и CRM] → [Автоматическое подтверждение заказа] → [Оптимизированная доставка]

- Автоматический прием заказа: CRM автоматически регистрирует заказ.
- Автоматическая обработка заказа и CRM: Заказ автоматически обрабатывается в CRM.
- Автоматическое подтверждение заказа: Клиент автоматически получает подтверждение заказа.
- Оптимизированная доставка: Логистика доставки оптимизирована для сокращения времени.
